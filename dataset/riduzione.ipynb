```python
import pandas as pd

# 1. Caricamento del dataset originale
# Sostituisci 'movies_dataset.csv' con il nome reale del tuo file
df = pd.read_csv('data.csv')

# 2. Riduzione all'ultimo decennio (1980 - 2023)
# NetworkX è spesso usato con grandi dataset, ma filtrare aiuta lo sviluppo rapido
df_ridotto = df[(df['Year of Release'] >= 1980) & (df['Year of Release'] <= 2023)]

# 3. Verifica dei risultati
print(f"Numero di film originali: {len(df)}")
print(f"Numero di film dopo la riduzione: {len(df_ridotto)}")

# 4. Salvataggio in un nuovo file CSV
df_ridotto.to_csv('dataset_ridotto.csv', index=False)

print("File 'dataset_ridotto.csv' salvato con successo!")
```
